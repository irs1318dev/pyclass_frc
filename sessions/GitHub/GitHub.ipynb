# GitHub
#### GitHub is a code hosting platform for version control and collaboration which lets you and others work together on projects together from anywhere.

#### To start off, set up a GitHub account at https://github.com/ 

In Git, you can create different repositories which are like folders for different projects
      - Repositories can contain folders and files of many kinds, images, videos, spreadsheets, data sets, etc.
      - You can get the repository onto your local machine using the 'git clone {repository URL}' command


Within your repository, it is good to add a README which is a short description of your repository
    - This helps others know what your repository entails before they start contributing

## Branches
By default, each repository has 1 branch known as main or master. You can create additional branches which are like copies of the main branch to code on before committing/pushing to the main branch; these branches run in parallel to the main branch and you can make branches off other branches. If people make changes to the main branch while you are working on a local branch, you can pull updates from the main branch to keep your code updated and avoid merge conflicts in the future. 

- To create a git branch through the command line, use 'git branch {branchname}'
- To checkout a git branch which is already made, use 'git checkout {branchname}'
- To do both tasks simultaneously, use 'git checkout -b {branchname}'
- To pull the most recent version of code from that branch, use 'git pull'
- To pull the most recent version of code from the main branch use 'git pull origin master'

## Adds, Commits, and Pushes
After you make changes to your code on your local branch, you need to add your changes to stage them for the next commit
- Use 'git add {filename}' to stage 1 file or 'git add .' to stage all tracked files


Once your files are staged, you can commit your files to your local branch
- Use 'git commit {filename}' to commit a particular file
- Use 'git commit -a' to commit all staged files
- Use 'git commit -a -m "Commit message"' to commit all staged files and include a commit message through the command line
When committing, you will always be required to enter a commit message. This is usually a concise description of changes you made within that commit for others to refer to when going through the changes in the code.


Once committed to your local repository, pushing pushes your branch to the remote repository
- Use 'git push' to push all committed files to the remote branch

## Pull Requests
When you open a pull request, youre proposing your changes and requesting that someone review and pull in your contribution and merge them into their branch. Pull requests show diffs, or differences, of the content from both branches. The changes, additions, and subtractions are shown in green and red. Once you are done looking over the differences and adding comments, you click the Create Pull Request button.

To merge these changes into your main branch, click the Merge pull request button. You can now delete your other branch since the changes have been incorporated in master!

## Merges without Pull Requests
Merging is used to combine multiple sequences of commits into one unified history. Most frequently, this is used to combine 2 branches or to decide what changes to keep when different people are working on the same code in the same branch at the same time. It takes two commit pointers, usually the branch tips, and finds a common base commit between them. Then, it creates a new "merge commit" that combines the changes of each queued merge commit sequence.


Usually, git can auto merge your changes but when the changes are made to the same lines of code, it creates merge conflicts which require user intervention to resolve

To merge a different branch into master:
- Commit your changes on your branch
- Checkout the master branch
- Use 'git merge {nameofbranchyouwanttomerge}' to merge your branch into master
    - This merges the specified branch into the currently active branch

## Other helpful commands
- 'git status' to look at what files you've changed
- 'git branch' to see what branch you are currently working on 
- 'git diff' shows the differences between commits, working trees, etc.
- 'git reset --hard' to unstage all the local commits
